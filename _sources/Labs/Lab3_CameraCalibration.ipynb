OpenMV Camera depth (the z-axis)

```python
y = np.arange(0.1, 0.28, 0.01) 
x = -np.array([5.23, 5.76, 6.25, 6.72, 7.25, 7.76, 8.25, 8.76, 9.30, 9.78, 10.31, 10.75, 11.33, 11.80, 12.36, 12.89, 13.40, 13.91])
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
print(m, c)
plt.plot(x, y, 'o', label='Measured data', markersize=10)
plt.plot(x, m*x + c, 'r', label='Fitted line')
plt.legend()
plt.grid('on')
plt.show()
```
